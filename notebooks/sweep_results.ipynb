# Sweep Results

Read the optimal hyperparamters found for the two sweeps: for muon as the optimizer for the convolutional layers and for sgd+momentum.

## Muon

```python
import wandb
from src.train_model import main
from types import SimpleNamespace

api = wandb.Api()

MUON_SWEEP_ID = "1fswjdht"

muon_sweep = api.sweep(f"adaptive-muon/muon_sweep/{MUON_SWEEP_ID}")

best_run_muon = muon_sweep.best_run()
best_config_muon = best_run_muon.config
best_config_muon = SimpleNamespace(**best_config_muon)

print("SGD+Momentum Optimal:")
print(f"Best Run Name: {best_run_muon.name}")
print(f"Optimal Hyperparameters: {best_config_muon}")
```

## SGD+Momentum

```python
SGD_MOMENTUM_SWEEP_ID = "l5iogopz"

sgd_sweep = api.sweep(f"adaptive-muon/muon_sweep/{SGD_MOMENTUM_SWEEP_ID}")

best_run_sgd = sgd_sweep.best_run()
best_config_sgd = best_run_sgd.config
best_config_sgd = SimpleNamespace(**best_config_sgd)

print("SGD+Momentum Optimal:")
print(f"Best Run Name: {best_run_sgd.name}")
print(f"Optimal Hyperparameters: {best_config_sgd}")
```

# Train models with optimal values

```python
# Muon
best_config_muon.val_split = 0.0
wandb.init(
     entity="adaptive-muon",
     project="cnn-training", 
     name="optimal_muon",
)
main(run="optimal_muon", 
     config=best_config_muon,
     device="mps",
     is_sweep=False)
wandb.finish()

# SGD + Momentum
best_config_sgd.val_split = 0.0
wandb.init(
     entity="adaptive-muon",
     project="cnn-training", 
     name="optimal_sgd",
)
main(run="optimal_sgd", 
     config=best_config_sgd,
     device="mps",
     is_sweep=False)
wandb.finish()
```
